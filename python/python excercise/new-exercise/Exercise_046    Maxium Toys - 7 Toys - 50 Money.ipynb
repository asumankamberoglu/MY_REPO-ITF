```python
n, m = list(map(int, input("Ltfen oyuncak adedi ve elinizdeki para miktarn girin: ").split()))
toys_prices = list(map(int, input("Ltfen oyuncak fiyatlarn girin: ").split()))
counter = 0
total = 0
for i in sorted(toys_prices) :
  if total + i <= m :
    counter += 1
    total += i
  else:
    break

print(f"{counter} adet oyuncak satn aldnz.")
```

```python
n, m = list(map(int, input("Ltfen oyuncak adedi ve elinizdeki para miktarn girin: ").split()))
toys_prices = list(map(int, input("Ltfen oyuncak fiyatlarn girin: ").split()))
counter = 0
total = 0
liste = []
for i in sorted(toys_prices) :
  if total + i <= m :
    counter += 1
    total += i
    liste.append(i)
  else:
    break
print("Fiyat {} olan oyuncaklar satn aldnz".format(liste))
print(f"{counter} adet oyuncak satn aldnz.")
```
