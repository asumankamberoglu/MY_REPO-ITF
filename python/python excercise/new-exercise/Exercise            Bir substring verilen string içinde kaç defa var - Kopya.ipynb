```python
string = input("Bir kelime yazınız: ")
sub_string= input("Bir alt kelime yazınız: ")
count = 0
for i in range(len(string)) :
  if string[i:i+len(sub_string)] == sub_string :
    count += 1
print(count)
```

```python
# Alternative

text = input("Enter a string: ")
sub = input("Enter a substring: ")
count = 0
for i in range(len(text)) :
  if text[i:].startswith(sub) :
    count += 1

print(count)
```

```python
age = input("Are you a cigarette addict older than 75 years old? (Yes/No) :").lower()
if age == "yes" :
  age = True
elif age == "no" :
  age = False

chronic = input("Do you have a chronic illness? (Yes/No): ").lower()
if chronic == "yes" :
  chronic = True
elif chronic == "no" :
  chronic = False

immune = input("Is your immune system strong? (Yes/No): ").lower()
if immune == "yes" :
  immune = False
elif immune == "no" :
  immune = True

risk = age or chronic or immune
if risk == True :
  print("You are in risky group")
else :
  print("You are not in risky group")
```
