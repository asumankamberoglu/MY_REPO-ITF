## `tell()` Metodu

```python
with open("fishes.txt", "r", encoding = "utf-8") as file :
    print(file.read(33))
    print(file.read(33))
    
    file.seek(0)
    print(file.read(33))
    print(file.tell())
```

```python
with open("istiklal.txt", "r", encoding="utf-8") as file:
    lines = file.readlines()
```

```python
lines
```

```python
counter = 0

with open("istiklal.txt", "w", encoding="utf-8") as file:
    for i in lines :
        counter +=1
        
        if counter % 4 == 0 :
            file.write(i + "\n")
            
        else : 
            file.write(i)
```
